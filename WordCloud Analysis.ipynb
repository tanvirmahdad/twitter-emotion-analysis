```python
import json
```

```python
# Here we will collect data from generated json file

wordlist=""
file = open('saturday.json', 'r')
for line in file:
    text = json.loads(line)
    words=text['text']
    wordlist+=words
file.close()

print(len(wordlist))
```

```python
#Import Libraries for wordcloud

import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline



import requests
from IPython.display import HTML


from bs4 import BeautifulSoup
import unidecode
import re
```

## Data Cleansing
Here, Data has been cleaned by removing links , backslash, Special characters from scrpaed tweets

```python
#Data Cleansing Steps

raw_text=wordlist

# Cleaning text by BeautifulSoup
soup = BeautifulSoup(raw_text)
all_tweets = ''.join(soup.findAll(text=True))

# Convert non-ascii characters into ASCII equivalent
all_tweets = unidecode.unidecode(all_tweets)

# # Remove backslash
all_tweets = all_tweets.replace("\'", "'")

#Remove Link
all_tweets = re.sub(r'http\S+', '', all_tweets)

#Remove all special characters
all_tweets = re.sub('[^A-Za-z ]+', ' ', all_tweets)

print(all_tweets)
```

## Word Cloud Generation

Now we are generating the Word Cloud

```python
# Here we are Generating the Word Cloud
#I have changed the background color so that words are more visible

param_wordcloud = {'max_font_size':20, 'max_words':80, 'background_color':"white"}
wordcloud = WordCloud(**param_wordcloud).generate(all_tweets)

# Display the generated image:
#increased the figuresize to fit in larger space
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
